```python
#3
#Simple Linear Regression
#Dataset Used - Income_happiness.csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the dataset
df = pd.read_csv("income.data.csv")
# Display the first few rows of the DataFrame
print(df.head())
# Prepare data for linear regression
X = df[['income']].values.reshape(-1, 1)
Y = df['happiness'].values 
#build Linear regression model
lm = LinearRegression()
#fit the model
lm.fit(X,Y)
#slope
lm.coef_

#intercept
lm.intercept_

lm.score(X,Y)
# Make predictions
Y_pred = lm.predict(X)
# Print the regression equation and parameters
slope = lm.coef_[0]
intercept = lm.intercept_
print(f"Linear Regression Equation: Happiness = {slope:.2f} * Income + {intercept:.2f}")
# Display the correlation coefficient
correlation = df.corr().loc['income', 'happiness']
print(f"Correlation coefficient: {correlation:.2f}")
# Scatter plot of the data
plt.scatter(df['income'], df['happiness'], color='orange', label='data')

# Plot the line of best fit
plt.plot(df['income'], Y_pred, color='black', linewidth=2, label='Line of Best Fit')
plt.title('Simple Linear Regression')
plt.xlabel('Income')
plt.ylabel('Happiness')
plt.legend()
plt.show()
# New income values for prediction
new_income = [[40000], [50000], [60000], [70000], [80000]]

# Predict the happiness for the new income values
predicted_happiness = lm.predict(new_income)

# Create a DataFrame for the new predictions
df_predicted = pd.DataFrame(new_income, columns=['income'])
df_predicted['Predicted_Happiness'] = predicted_happiness

# Save the predicted results to a CSV file
df_predicted.to_csv('D:/230970034_ML/predicted.csv', index=False)
```

```python
#4)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the data from the CSV files
income_data = pd.read_csv(r'D:\230970034_ML\income.data.csv')
salaries_data = pd.read_csv(r'D:\230970034_ML\Salaries.csv')

# Linear Regression on Income vs Happiness
X_income = income_data[['income']].values
y_happiness = income_data['happiness'].values

model_income = LinearRegression()
model_income.fit(X_income, y_happiness)
y_pred_income = model_income.predict(X_income)

plt.figure(figsize=(8, 6))
plt.scatter(X_income, y_happiness, color='blue', label='Actual data')
plt.plot(X_income, y_pred_income, color='red', label='Regression line')
plt.title('Linear Regression: Income vs Happiness')
plt.xlabel('Income')
plt.ylabel('Happiness')
plt.legend()
plt.grid(True)
plt.show()

X_service = salaries_data[['yrs.service']].values
y_salary = salaries_data['salary'].values

model_salary = LinearRegression()
model_salary.fit(X_service, y_salary)
y_pred_salary = model_salary.predict(X_service)

plt.figure(figsize=(8, 6))
plt.scatter(X_service, y_salary, color='green', label='Actual data')
plt.plot(X_service, y_pred_salary, color='red', label='Regression line')
plt.title('Linear Regression: Years of Service vs Salary')
plt.xlabel('Years of Service')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()


print(f"Income vs Happiness Coefficient: {model_income.coef_[0]}, Intercept: {model_income.intercept_}")
print(f"Years of Service vs Salary Coefficient: {model_salary.coef_[0]}, Intercept: {model_salary.intercept_}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

abalone = pd.read_csv(r'D:/230970034_ML/abalone.csv')

# Extract features and target variable
X_abalone = abalone[['Length']].iloc[:20].values
y_abalone = abalone['Rings'].iloc[:20].values

# .iloc[:20].values

# Initialize and fit the linear regression model
model_abalone = LinearRegression()
model_abalone.fit(X_abalone, y_abalone)

slope = model_abalone.coef_[0]
intercept = model_abalone.intercept_

y_pred = model_abalone.predict(X_abalone)

r2 = r2_score(y_abalone, y_pred)
sse = np.sum((y_abalone - y_pred) ** 2)
sst = np.sum((y_abalone - np.mean(y_abalone)) ** 2)

print(f'Linear Regression Equation: y = {slope:.2f}x + {intercept:.2f}')
print(f'Final Parameter Values: Slope = {slope:.2f}, Intercept = {intercept:.2f}')
print(f'Predicted Values (y^): {y_pred}')
print(f'R-squared: {r2:.2f}')
print(f'Sum of Squared Errors (SSE): {sse:.2f}')
print(f'Total Sum of Squares (SST): {sst:.2f}')

plt.figure(figsize=(10, 6))
plt.scatter(X_abalone, y_abalone, color='blue', label='Data points')
plt.plot(X_abalone, y_pred, color='red', linewidth=2, label='Line of Best Fit')
plt.title('Linear Regression: abalone Dataset')
plt.xlabel('Length')
plt.ylabel('Rings')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load mtcars dataset
df_mtcars = pd.read_csv(r'mtcars.csv')

# Extract X (horsepower) and Y (miles per gallon)
X_mtcars = df_mtcars[['hp']].values
Y_mtcars = df_mtcars['mpg'].values

# Initialize and fit the linear regression model
model_mtcars = LinearRegression()
model_mtcars.fit(X_mtcars, Y_mtcars)

# Get the slope (m) and intercept (b) of the line
slope = model_mtcars.coef_[0]
intercept = model_mtcars.intercept_


# Predict Y values
Y_pred_mtcars = model_mtcars.predict(X_mtcars)

r2 = r2_score(Y_mtcars, Y_pred_mtcars)
sse = np.sum((Y_mtcars - Y_pred_mtcars) ** 2)
sst = np.sum((Y_mtcars - np.mean(Y_pred_mtcars)) ** 2)

print(f'Linear Regression Equation: y = {slope:.2f}x + {intercept:.2f}')
print(f'Final Parameter Values: Slope = {slope:.2f}, Intercept = {intercept:.2f}')
print(f'Predicted Values (y^): {y_pred}')
print(f'R-squared: {r2:.2f}')
print(f'Sum of Squared Errors (SSE): {sse:.2f}')
print(f'Total Sum of Squares (SST): {sst:.2f}')

# Plot the data points and the line of best fit
plt.figure(figsize=(10, 6))
plt.scatter(X_mtcars, Y_mtcars, color='blue', label='Data points')
plt.plot(X_mtcars, Y_pred_mtcars, color='red', linewidth=2, label='Line of Best Fit')
plt.title('Linear Regression: mtcars Dataset')
plt.xlabel('Horsepower (hp)')
plt.ylabel('Miles per Gallon (mpg)')
plt.legend()
plt.grid(True)
plt.show()
```
