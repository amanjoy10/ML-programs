```python
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
x=data.data
y=data.target
```

```python
from sklearn.model_selection import train_test_split as tts
xtrn,xtst,ytrn,ytst=tts(x,y,test_size=.2)
```

```python
from sklearn import svm

def model(kernel="linear",c=1.0,gama='auto'):
    if kernel=='rbf':
        cls=svm.SVC(kernel=kernel,C=c,gamma=gama)
    else:
        cls=svm.SVC(kernel=kernel,C=c)
    cls.fit(xtrn,ytrn)
    return (cls.predict(xtst),cls.predict(xtrn))
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import seaborn as sns
import matplotlib.pyplot as plt
def metric(yptst,yptrn):
    print(f'Accuracy Score Test: {accuracy_score(ytst,yptst)}')
    print(f'Accuracy Score Train: {accuracy_score(ytrn,yptrn)}')
    print(f'classification report: {classification_report(ytst,yptst)}')
    cm=confusion_matrix(ytst,yptst)
    print(f'Confusion matrix: {cm}')
    sns.heatmap(cm)
    sns.heatmap(cm,annot=True,cmap='YlGnBu',xticklabels=['Belgn','malgin'],yticklabels=['Belgn','Malgin'])
    plt.show()
```

```python
kernel=['linear','rbf','sigmoid','poly']
c=[1.0,100.0,1000.0]

for k in kernel:
    for i in c:
        print(f'Kernel: {k}\t C: {i}')
        yptst,yptrn=model(k,i)
        metric(yptst,yptrn)
```
